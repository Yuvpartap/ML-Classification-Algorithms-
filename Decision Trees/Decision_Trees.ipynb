```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weatherAUS.csv')
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['RainTomorrow'], inplace=True)
```

```python
df['year'] = pd.to_datetime(df['Date']).dt.year
```

```python
df.head()
```

```python
df.drop(['Date'], axis = 1, inplace = True)
df.head()
```

```python
train_data = df[df['year']<2015]
val_data = df[df['year']==2015]
test_data = df[df['year']>2015]
```

```python
print(train_data.shape)
print(val_data.shape)
print(test_data.shape)
```

```python
input_cols = list(train_data.columns)[0:-2]
target_cols = ['RainTomorrow']
```

```python
input_cols
```

```python
train_input = train_data[input_cols].copy()
val_input = val_data[input_cols].copy()
test_input = test_data[input_cols].copy()

train_target = train_data[target_cols].copy()
val_target = val_data[target_cols].copy()
test_target = test_data[target_cols].copy()
```

```python
# Numerical and categorical columns
num_col = train_input.select_dtypes(include=np.number).columns.tolist()
cat_col = train_input.select_dtypes('object').columns.tolist()
```

```python
print("numerical columns : ",num_col)
print("categorical columns : ",cat_col)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean').fit(df[num_col])
```

```python
train_input[num_col] = imputer.transform(train_input[num_col])
val_input[num_col] = imputer.transform(val_input[num_col])
test_input[num_col] = imputer.transform(test_input[num_col])
```

```python
train_input.head()
```

```python
#Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(df[num_col])

                            
```

```python
train_input[num_col] = scaler.transform(train_input[num_col])
test_input[num_col] = scaler.transform(test_input[num_col])
val_input[num_col] = scaler.transform(val_input[num_col])
train_input.describe().loc[['min','max']]
```

```python
#ENCODING DATA
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False, handle_unknown="ignore").fit(df[cat_col])
encoded_cols =list(encoder.get_feature_names(cat_col))
encoded_cols
```

```python
train_input[encoded_cols] = encoder.transform(train_input[cat_col])
val_input[encoded_cols] = encoder.transform(val_input[cat_col])
test_input[encoded_cols] = encoder.transform(test_input[cat_col])
train_input
```

```python
x_train = train_input[num_col + encoded_cols]
x_test = test_input[num_col + encoded_cols]
x_val = val_input[num_col + encoded_cols]
x_train
```

```python
#Training Data
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state = 30)
model.fit(x_train, train_target)
```

```python
predics= model.predict(x_train)
pd.value_counts(predics)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(predics, train_target)
```

```python
probability= model.predict_proba(x_train)
probability
```

```python
model.score(x_val,val_target)
```
